```python
from u11link import U11linkModel as u11
from u11link import ObservablesU11 as obs
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-np.pi, np.pi, 100)
y = [u11().bosonic_action(0.3, p) for p in x]

plt.plot(x, y)
plt.xlabel(r"x")
plt.ylabel(r"y")
plt.title(r"Bosonic action at $\beta = 0.3$")
plt.show()
```

```python
mu = np.linspace(0, 8, 100)
k = 0.5
model = u11()
```

## Polyakov loop

```python
y_1 = [model.expectation_value(beta=1, mu=m, k=k, operator=obs.polyakov_loop) for m in mu]
y_2 = [model.expectation_value(beta=2, mu=m, k=k, operator=obs.polyakov_loop) for m in mu]
y_3 = [model.expectation_value(beta=3, mu=m, k=k, operator=obs.polyakov_loop) for m in mu]

y_1_exact = [obs.exp_polyakov_loop(beta=1, mu=m, k=k) for m in mu]
y_2_exact = [obs.exp_polyakov_loop(beta=2, mu=m, k=k) for m in mu]
y_3_exact = [obs.exp_polyakov_loop(beta=3, mu=m, k=k) for m in mu]


plt.scatter(mu, y_1, label=r"$\beta = 1$", s=5)
plt.scatter(mu, y_2, label=r"$\beta = 2$", s=5)
plt.scatter(mu, y_3, label=r"$\beta = 3$", s=5)

plt.plot(mu, y_1_exact, label=r"$\beta = 1$ exact")
plt.plot(mu, y_2_exact, label=r"$\beta = 2$ exact")
plt.plot(mu, y_3_exact, label=r"$\beta = 3$ exact")

plt.xlabel(r"$\mu$")
plt.ylabel(r"$\mathbb{R}e(<e^{ix}>)$")
plt.title(r"$\mathbb{R}e(<e^{ix}>)$ vs $\mu$, with $\kappa = 0.5$")
plt.legend()
plt.show()
```

## Inverse Polyakov loop

```python
mu = np.linspace(0, 8, 100)
k = 0.5
model = u11()

y_1 = [model.expectation_value(beta=1, mu=m, k=k, operator=obs.inverse_polyakov_loop) for m in mu]
y_2 = [model.expectation_value(beta=2, mu=m, k=k, operator=obs.inverse_polyakov_loop) for m in mu]
y_3 = [model.expectation_value(beta=3, mu=m, k=k, operator=obs.inverse_polyakov_loop) for m in mu]

y_1_exact = [obs.exp_inverse_polyakov_loop(beta=1, mu=m, k=k) for m in mu]
y_2_exact = [obs.exp_inverse_polyakov_loop(beta=2, mu=m, k=k) for m in mu]
y_3_exact = [obs.exp_inverse_polyakov_loop(beta=3, mu=m, k=k) for m in mu]


plt.scatter(mu, y_1, label=r"$\beta = 1$", s=5)
plt.scatter(mu, y_2, label=r"$\beta = 2$", s=5)
plt.scatter(mu, y_3, label=r"$\beta = 3$", s=5)

plt.plot(mu, y_1_exact, label=r"$\beta = 1$ exact")
plt.plot(mu, y_2_exact, label=r"$\beta = 2$ exact")
plt.plot(mu, y_3_exact, label=r"$\beta = 3$ exact")

plt.xlabel(r"$\mu$")
plt.ylabel(r"$\mathbb{R}e(<e^{ix}>)$")
plt.title(r"$\mathbb{R}e(<e^{ix}>)$ vs $\mu$, with $\kappa = 0.5$")
plt.legend()
plt.show()
```

## Plaquette

```python
mu_i = 1j * np.linspace(-8, 0, 100)
mu_r = np.linspace(0, 8, 100)

mu = np.concatenate((mu_i, mu_r))

mu_squared = mu**2

y_1 = [model.expectation_value(beta=1, mu=m, k=k, operator=obs.plaquette) for m in mu]
y_2 = [model.expectation_value(beta=2, mu=m, k=k, operator=obs.plaquette) for m in mu]
y_3 = [model.expectation_value(beta=3, mu=m, k=k, operator=obs.plaquette) for m in mu]

y_1_exact = [obs.exp_plaquette(beta=1, mu=m, k=k) for m in mu]
y_2_exact = [obs.exp_plaquette(beta=2, mu=m, k=k) for m in mu]
y_3_exact = [obs.exp_plaquette(beta=3, mu=m, k=k) for m in mu]


plt.scatter(mu_squared, y_1, label=r"$\beta = 1$", s=5)
plt.scatter(mu_squared, y_2, label=r"$\beta = 2$", s=5)
plt.scatter(mu_squared, y_3, label=r"$\beta = 3$", s=5)

plt.plot(mu_squared, y_1_exact, label=r"$\beta = 1$ exact")
plt.plot(mu_squared, y_2_exact, label=r"$\beta = 2$ exact")
plt.plot(mu_squared, y_3_exact, label=r"$\beta = 3$ exact")

plt.xlabel(r"$\mu^2$")
plt.ylabel(r"$\mathbb{R}e(<cos(x)>)$")
plt.title(r"$\mathbb{R}e(<cos(x)>)$ vs $\mu^2$, with $\kappa = 0.5$")
plt.legend()
plt.show()
```
