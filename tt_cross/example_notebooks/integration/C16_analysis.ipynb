```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

```python
mc16 = pd.read_csv("C12_data/C12_mc.csv")
gr16 = pd.read_csv("C12_data/C12_greedy.csv", header=None, names=["n_samples", "integral"])
exact = 0.6304735033743867961220401
plt.figure(figsize=(10, 6), dpi=200)
plt.plot(mc16["n_samples"][100:], np.abs(mc16["integral"] - exact)[100:], label="MC")
plt.plot(gr16["n_samples"], np.abs(gr16["integral"] - exact), label="Greedy")
plt.xlabel("Number of calls to $f(x_1,...,x_{16})$")
plt.legend()
plt.yscale("log")
plt.ylabel("Error")
plt.show()
```
